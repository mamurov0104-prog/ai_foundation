```python
import torch
from torch import nn # PyTorch blocks
import matplotlib.pyplot as plt
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"the device: {device}")
```

```python
# create weight, bias and range
weight = 0.7
bias = 0.3

start = 0
end = 1
step = 0.02

# 2D bolishi kutiladi yani qiymat tashqari unsqueeze orqali feature berdik
X = torch.arange(start, end, step).unsqueeze(dim=1)
print("X:", X, "\n")
# Linier Regression > Y = wX+b
Y = weight * X + bias
print("Y:", Y, "\n")

# Split data > TRAIN data vs TEST data
train_split = int(0.8 * len(X))
x_train, y_train = X[:train_split], Y[:train_split]
x_test, y_test = X[train_split:], Y[train_split:]

len(x_train), len(y_train), len(x_test), len(y_test)

```

```python
def plot_predictions(train_data=x_train,
                     train_labels=y_train,
                     test_data=x_test,
                     test_labels=y_test,
                     predictions=None):

    plt.figure(figsize=(10, 7))

     # Plot training data in blue
    plt.scatter(train_data, train_labels, c='b', s=25, label='Training data')

    # Plot test data in green
    plt.scatter(test_data, test_labels, c='g', s=25, label='Testing data')

    if predictions is not None:
        plt.scatter(test_data, predictions, c='r', s=25, label='Predictions')

    # Styling
    plt.title("Linear Regression Model Predictions", fontsize=16)
    plt.xlabel("X values", fontsize=14)
    plt.ylabel("Y values", fontsize=14)
    plt.legend(fontsize=12)

    plt.tight_layout()
    plt.show()

plot_predictions()
```

```python
# Parent nn.Module inherintence olamiz
class LinearRegressionModelV2(nn.Module):
    def __init__(self):
        super().__init__()
        # nn.Linear() orqali model ichida layer yaratamix
        # weight va bias avtomatik hosil etiladi
        self.linear_layer = nn.Linear(in_features=1,
                                      out_features=1)

    # forward computation belgilash (input data x flows through nn.Linear())
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)


# Model hosil bolishida har safar bir xil Random weightlar chiqishi uchun
torch.manual_seed(42)

model_1 = LinearRegressionModelV2() # Model yaratish
model_1, model_1.state_dict() # Model ichidagi barcha parametrlarni ko‘rish
```

```python
# Loss Function | MAE => Mean Absolute Error
loss_fn = nn.L1Loss()

# Optimizer - biz hohlagan weight va biasga yaqinlashgan holatga borib tohtaydi
optimizer = torch.optim.SGD(
    params=model_1.parameters(),
    lr=0.01 # BackPropagationdagi learning rate
    # Forward → Loss → Backward → Gradient * lr → Optimizer → modify Weight
)
```

```python
# See model's untrained predictions (on test set)
model_1.eval()

with torch.inference_mode():
    y_preds_early = model_1(x_test)  # backwards (Back Propagation)

plot_predictions(predictions=y_preds_early)
plt.title("Predictions before training (red = random guesses)")
```

```python
# Set the number of epochs
epochs = 1000

# Put data on the available device
# Without this, error will happen (not all model/data on device)
x_train = x_train.to(device)
x_test = x_test.to(device)
y_train = y_train.to(device)
y_test = y_test.to(device)

for epoch in range(epochs):
    ### Training
    model_1.train()  # train mode is on by default after construction

    # 1. Forward pass
    y_pred = model_1(x_train)

    # 2. Calculate loss
    loss = loss_fn(y_pred, y_train)

    # 3. Zero gradient optimizer
    optimizer.zero_grad()

    # 4. Loss backward
    loss.backward()

    # 5. Step the optimizer
    optimizer.step()

    ### Testing
    model_1.eval()  # put the model in evaluation mode for testing (inference)

    # 1. Forward pass
    with torch.inference_mode():
        test_pred = model_1(x_test)

        # 2. Calculate the loss
        test_loss = loss_fn(test_pred, y_test)

    if epoch % 100 == 0:
        print(f"Epoch: {epoch} | Train loss: {loss} | Test loss: {test_loss}")
```

```python
# Turn model into evaluation mode
model_1.eval()

# Make predictions on the test data
with torch.inference_mode():
    y_preds = model_1(x_test)

y_preds
```

```python
plot_predictions(predictions=y_preds)
plt.title("Predictions After training (red = random guesses)")
```
